````python
raw_df['DateTime'] = raw_df['DateTime'].astype('datetime64[ns]')
raw_df['weekday'] = raw_df['DateTime'].dt.weekday
raw_df['year'] = raw_df['DateTime'].dt.year
raw_df['month'] = raw_df['DateTime'].dt.month
raw_df['day'] = raw_df['DateTime'].dt.day
original_df = raw_df.copy()
````

````python
raw_df
````

````python
raw_df['weekday'].replace(0,'01 - Monday',inplace=True)
raw_df['weekday'].replace(1,'02 - Tuesday',inplace=True)
raw_df['weekday'].replace(2,'03 - Wednesday',inplace=True)
raw_df['weekday'].replace(3,'04 - Thursday',inplace=True)
raw_df['weekday'].replace(4,'05 - Friday',inplace=True)
raw_df['weekday'].replace(5,'06 - Saturday',inplace=True)
raw_df['weekday'].replace(6,'07 - Sunday',inplace=True)

raw_df['ID'].nunique()
raw_df
````

````python
train_data = raw_df.groupby(['month','weekday'])['Vehicles'].mean().reset_index()

train_data = train_data.pivot('weekday','month','Vehicles')
train_data.sort_index(inplace=True)

sns.set(font_scale=1.2) 

f, ax = plt.subplots(figsize=(16, 8))
sns.heatmap(train_data, annot=False, ax=ax, fmt="d", linewidths=2)
plt.title('Web Traffic Months cross Weekdays')
plt.show()
````

````python
import numpy as np
def rmse(targets,predictions):
    return np.sqrt(np.mean(np.square(targets-predictions)))  
````

````python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
inputs, targets = raw_df[['month','year','day','Junction']], raw_df['Vehicles']
model = LinearRegression().fit(inputs,targets)
predictions = model.predict(inputs)

loss = rmse(targets,predictions)
print('Loss',loss)


````
